# Numpy

```python
import numpy as np
```

```python
A = np.array([[1, 2, -4],
              [-5, 0, 1],
              [3, 2, 3]])

B = np.array([[-4, 0, 3],
              [1, 0, 2],
              [-4, 1, 8]])
```

```python
detA = np.linalg.det(A)
detB = np.linalg.det(B)
print(detA)
print(detB)
```

```python
invA = np.linalg.inv(A)
invB = np.linalg.inv(B)

print(invA)
print()
print(invB)
```

```python
res = A @ B
res
```

```python
res2 = (B.T @ A.T).T
res2
```

### Pyplot

```python
import matplotlib.pyplot as plt
```

```python
x_sinh = np.linspace(-6, 0, 100)
x_cosh = np.linspace(-6, 5, 200)
x_f = np.linspace(7, 2, 300)
x = np.linspace(-4, 4, 600)

y_sinh = np.sinh(x_sinh)
y_cosh = np.cosh(x_cosh)
y_f = x_f**3 + x_f**2 + x_f + 1
y = np.exp(x) + np.exp(-x)

plt.plot(x_sinh, y_sinh, c= "red", label = "sinh")
plt.plot(x_cosh, y_cosh, c= "blue", label = "cosh")
plt.plot(x_f, y_f, c= "orange", label = "funcion")
plt.plot(x, y, c= "black", label = "Exp")

plt.title('Funciones')
plt.grid(True)
plt.legend()
plt.show()
```
