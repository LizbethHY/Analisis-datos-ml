```python
def media(lista, k):
    res = []
    i = 0
    while i < len(lista):
        suma = 0
        cantidad = 0
        j = 0
        while j < k and i + j < len(lista):
            suma += lista[i + j]
            cantidad += 1
            j += 1
        promedio = suma / cantidad
        res.append(promedio)
        i += k
    return res
```

```python
lista = [1, 2, 3, 4, 5, 6, 7, 8]

print(media(lista, 3))
print(media(lista, 2))
```

```python
def primos(n):
    res = []
    for num in range(2, n + 1):
        primo = True
        for i in range(2, int(num**1/2) + 1):
            if num % i == 0:
                primo = False
                break
        if primo:
            res.append(num)
    print(res)
```

```python
primos(9)
primos(29)
```

```python
def contar(cadena):
    acumulador = {}
    for letra in cadena:
        if letra in acumulador:
            acumulador[letra] += 1
        else:
            acumulador[letra] = 1
    for letra, cant in sorted(acumulador.items()):
        print(f"{letra}: {cant}")
```

```python
contar("ABAADCBCD")
print("\n")
contar("AEEEAAIIAOOUUAEU")
```

```python
unidades = ["", "uno", "dos", "tres", "cuatro", "cinco", "seis", "siete", "ocho", "nueve", "diez", "once",
            "doce", "trece", "catorce", "quince", "dieciséis", "diecisiete", "dieciocho", "diecinueve"]
decenas = ["", "diez", "veinte", "treinta", "cuarenta", "cincuenta", "sesenta", "setenta", "ochenta", "noventa"]
centenas = ["", "ciento", "doscientos", "trescientos", "cuatrocientos", "quinientos", "seiscientos",
            "setecientos", "ochocientos", "novecientos"]
```

```python
def literal(n):
    if (n<= 0 and n > 1000000):
        return "Número fuera de rango"
    if n == 0:
        return "cero"
    if n == 1000000:
        return "un millón"
    if n == 100:
        return "cien"
    if n < 20:
        return unidades[n]
    if n < 30:
        return "veinte" if n == 20 else "veinti" + unidades[n - 20]
    if n < 100:
        return decenas[n // 10] if n % 10 == 0 else decenas[n // 10] + " y " + unidades[n % 10]
    if n < 1000:
        return centenas[n // 100] if n % 100 == 0 else centenas[n // 100] + " " + literal(n % 100)
    if n < 2000:
        return "mil" if n % 1000 == 0 else "mil " + literal(n % 1000)
    return (literal(n // 1000) + " mil" if n % 1000 == 0 else literal(n // 1000) + " mil " + literal(n % 1000))
```

```python
print(literal (9))
print(literal (24))
print(literal (123))
print(literal (3246))
print(literal (583345))
print(literal (1000000))
```

```python
class Persona:
    def __init__(self, id, nombre, apellido):
        self._id = id
        self._nombre = nombre
        self._apellido = apellido

    @property
    def id(self):
        return self._id

    @id.setter
    def id(self, nuevo_id):
        self._id = nuevo_id

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, nuevo_nombre):
        self._nombre = nuevo_nombre

    @property
    def apellido(self):
        return self._apellido

    @apellido.setter
    def apellido(self, nuevo_apellido):
        self._apellido = nuevo_apellido

    def hablar(self):
        print('Hola mundo...')

    def caminar(self):
        print('Caminando...')

    def mostrar_datos_personales(self):
        print(f"ID: {self._id}, Nombre: {self._nombre}, Apellido: {self._apellido}")
```

```python
class Docente(Persona):
  def __init__ (self, id,nombre, apellido, salario, fecha_ingreso):
    super().__init__(id, nombre, apellido)
    self.__salario = salario
    self.__fecha_ingreso = fecha_ingreso

  @property
  def salario(self):
      return self.__salario

  @property
  def fecha_ingreso(self):
      return self.__fecha_ingreso

  @salario.setter
  def salario(self, nuevo_salario):
      self.__salario = nuevo_salario

  @fecha_ingreso.setter
  def fecha_ingreso(self, nueva_fecha_ingreso):
      self.__fecha_ingreso = nueva_fecha_ingreso

  def dictar_clases(self):
        print('Dictando clases...')

  def __str__(self):
        return f"{self._id},{self._nombre},{self._apellido},{self.__salario},{self.__fecha_ingreso}"

  def mostrar_datos_personales(self):
        print(f"ID: {self._id}, Nombre: {self._nombre}, Apellido: {self._apellido}, Nivel: {self.__salario}, Fecha ingreso: {self.__fecha_ingreso}")
```
